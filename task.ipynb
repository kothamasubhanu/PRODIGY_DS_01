```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
import os

# Change the working directory to C:\jupyter
os.chdir('C:\\jupyter')

# Verify the current working directory
print(os.getcwd())
```

```python
# Load the dataset
file_path = r"C:\\jupyter\API_SP.POP.TOTL_DS2_en_csv_v2_342248\API_SP.POP.TOTL_DS2_en_csv_v2_342248.csv"
data = pd.read_csv(file_path, skiprows=4)  # Skip the first 4 rows for this dataset

# Display the first few rows of the dataframe
data.head()
print(data.info())

# Get summary statistics
print(data.describe())

```

```python
# Select relevant columns
population_data = data[['Country Name', '2022']]  # Use the most recent year available

# Drop rows with missing values
population_data.dropna(inplace=True)

# Display the cleaned data
population_data.head()
```

```python
# Plot the histogram
plt.figure(figsize=(10, 5))
plt.hist(population_data['2022'], bins=30, edgecolor='black')
plt.title('Population Distribution by Country in 2022')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.show()
```

```python
# Sort data by population and select top 10
top_10_countries = population_data.sort_values(by='2022', ascending=False).head(10)

# Plot the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries['Country Name'], top_10_countries['2022'], color='skyblue')
plt.title('Top 10 Most Populated Countries in 2022')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```
